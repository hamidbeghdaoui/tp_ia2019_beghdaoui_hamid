```python

import pandas as pd 
import numpy as nb
data = pd.read_csv('breastCancer (2).csv')
data
```

```python
data.isnull().sum()
```

```python
data['bare_nucleoli'].replace(['?'], 0,inplace=True)
data['class'].replace([2,4], [0,1],inplace=True)
```

```python

X=data.iloc[:,:-1]
Y=data['class']
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.4,stratify=Y)
```

```python
from sklearn.model_selection import KFold
```

```python

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

grid_params = {
    'n_neighbors':[3,5,7,9,11],
    'metric':['euclidean','manhattan']
}

gs = GridSearchCV(
    KNeighborsClassifier(),
    grid_params,
    verbose = 1,
    cv = 3,
    n_jobs=-1
)
gs_results = gs.fit(x_train,y_train)
```

```python
gs_results.best_params_
```

```python
#algo de knnv 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9, metric='euclidean')
knn.fit(x_train, y_train)
R= knn.predict(x_test)
```

```python
from sklearn.metrics import precision_score, recall_score
print('precision:',precision_score(y_test, R) )
print('recall:',recall_score(y_test, R) )
```

```python
from sklearn.metrics import  confusion_matrix
confusion_matrix(y_test, R)
```
